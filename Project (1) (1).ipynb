```python
def set_month(row):
    if row["month"] == 'ม.ค.':
        return "January"
    elif row["month"] == 'ก.พ.':
        return "February"
    elif row["month"] == 'มี.ค.':
        return "March"
    elif row["month"] == 'เม.ย.':
        return "April"
    elif row["month"] ==  'พ.ค.':
        return "May"
    elif row["month"] == 'มิ.ย.':
        return "June"
    elif row["month"] == 'ก.ค.':
        return "July"
    elif row["month"] == 'ส.ค':
        return "August"
    elif row["month"] == 'ก.ย':
        return "September"
    elif row["month"] ==  'ต.ค.':
        return "October"
    elif row["month"] == 'พ.ย.':
        return "November"
    elif row["month"] == 'ธ.ค.':
        return "DEcember"
    else:
        return "ไม่มี"
    
df = df.assign(month=df.apply(set_month, axis=1))
     
```
